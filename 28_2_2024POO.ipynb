```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def presentar(self):
        return f"Me llamo {self.nombre} y tengo {self.edad} años."

class Estudiante(Persona):
    def __init__(self, nombre, edad, grado):
        super().__init__(nombre, edad)
        self.__grado = grado

    def get_grado(self):
        return self.__grado

    def set_grado(self, nuevo_grado):
        self.__grado = nuevo_grado

    def estudiar(self, materia):
        return f"{self.nombre} está estudiando {materia}."

    def __str__(self):
        return f"la Estudiante: {self.nombre} tiene  {self.edad} años y esta en el Grado: {self.__grado}"


estudiante1 = Estudiante("flor", 14, "10° grado")


print(estudiante1.presentar())
print(estudiante1.estudiar("ingles"))
print(estudiante1.get_grado())


estudiante1.set_grado("11° grado")
print(estudiante1.get_grado())


print(estudiante1)
```

```python
class Animal:
    def __init__(self, nombre, especie):
        self.nombre = nombre
        self.especie = especie

    def emitir_sonido(self):
        pass

class Elefante(Animal):
    def emitir_sonido(self):
        return "Barritar!"


class Leon(Animal):
    def emitir_sonido(self):
        return "¡Rugir!"


elefante = Elefante("Dumbo", "Elefante")
leon = Leon("Simba", "León")


print(elefante.emitir_sonido())
print(leon.emitir_sonido())
```

# CUERPOS CELESTES

```python
class CuerpoCeleste:
    def __init__(self, nombre, tipo):
        self.nombre = nombre
        self.tipo = tipo


class Estrella(CuerpoCeleste):
    def __init__(self, nombre, tipo, luminosidad):
        super().__init__(nombre, tipo)
        self.luminosidad = luminosidad


class Planeta(CuerpoCeleste):
    def __init__(self, nombre, tipo, sistema):
        super().__init__(nombre, tipo)
        self.sistema = sistema


estrella = Estrella("Sirio", "Estrella", "Brillante")
planeta = Planeta("Marte", "Planeta", "Sistema Solar")

print(estrella.nombre, estrella.luminosidad)
print(planeta.nombre, planeta.sistema)
```

```python
class CuerpoCeleste:
    def __init__(self, nombre, tipo):
        self.nombre = nombre
        self.tipo = tipo


class Estrella(CuerpoCeleste):
    def __init__(self, nombre, tipo, luminosidad):
        super().__init__(nombre, tipo)
        self.luminosidad = luminosidad


class Planeta(CuerpoCeleste):
    def __init__(self, nombre, tipo, sistema):
        super().__init__(nombre, tipo)
        self.sistema = sistema


class SateliteNatural(CuerpoCeleste):
    def __init__(self, nombre, tipo, planeta):
        super().__init__(nombre, tipo)
        self.planeta = planeta


estrella = Estrella("Sirio", "Estrella", "Brillante")
planeta = Planeta("Marte", "Planeta", "Sistema Solar")
satelite = SateliteNatural("Luna", "Satélite Natural", "Tierra")

print(estrella.nombre, estrella.luminosidad)
print(planeta.nombre, planeta.sistema)
print(satelite.nombre, satelite.planeta)
```

# BEBIDAS

```python
class Bebida:
    def __init__(self, nombre, tipo):
        self.nombre = nombre
        self.tipo = tipo


class JugoNaranja(Bebida):
    def __init__(self, nombre, tipo, sabor):
        super().__init__(nombre, tipo)
        self.sabor = sabor


class Agua(Bebida):
    def __init__(self, nombre, tipo, mineral):
        super().__init__(nombre, tipo)
        self.mineral = mineral


class Cafe(Bebida):
    def __init__(self, nombre, tipo, intensidad):
        super().__init__(nombre, tipo)
        self.intensidad = intensidad


class Refresco(Bebida):
    def __init__(self, nombre, tipo, sabor):
        super().__init__(nombre, tipo)
        self.sabor = sabor


class Te(Bebida):
    def __init__(self, nombre, tipo, variedad):
        super().__init__(nombre, tipo)
        self.variedad = variedad


class Leche(Bebida):
    def __init__(self, nombre, tipo, origen):
        super().__init__(nombre, tipo)
        self.origen = origen


jugo_naranja = JugoNaranja("Jugo de Naranja", "Jugo", "Naranja")
agua = Agua("Agua", "No alcohólica", "Sin gas")
cafe = Cafe("Café", "Caliente", "Fuerte")
refresco = Refresco("Coca-Cola", "Refresco", "Cola")
te = Te("Té", "Caliente", "Negro")
leche = Leche("Leche", "Láctea", "Vaca")


print(jugo_naranja.nombre, jugo_naranja.sabor)
print(agua.nombre, agua.mineral)
print(cafe.nombre, cafe.intensidad)
print(refresco.nombre, refresco.sabor)
print(te.nombre, te.variedad)
print(leche.nombre, leche.origen)
```
